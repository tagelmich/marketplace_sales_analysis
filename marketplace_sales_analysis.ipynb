# Импортируем и предобрабатываем данные

```python
import pandas as pd
import requests

customers_link = "https://disk.yandex.ru/d/yN7JhG1QCmR31g"
items_link = "https://disk.yandex.ru/d/AoUAKlX2xR6p2w"
orders_link = "https://disk.yandex.ru/d/0w3EVr3L-t0MSA"


def load_csv_by_public_link(public_link, sep=','):

    response = requests.get(
        "https://cloud-api.yandex.net/v1/disk/public/resources/download",
        params={"public_key": public_link}
    )
    if response.status_code != 200:
        raise Exception(f"Ошибка: {response.status_code}, {response.text}")
    
    download_url = response.json()["href"]
    return pd.read_csv(download_url, sep=sep)




customers = load_csv_by_public_link(customers_link)
items = load_csv_by_public_link(items_link)
orders = load_csv_by_public_link(orders_link)
```

```python

orders_by_customers['order_period']=orders_by_customers['order_purchase_timestamp'].dt.to_period('M')
orders_by_customers['order_period_day']=orders_by_customers['order_purchase_timestamp'].dt.to_period('D')
orders_by_customers['order_difference']=orders_by_customers['order_purchase_timestamp']-orders['order_purchase_timestamp']
```

```python


#берем только те заказы, у которых статус "delivered"

orders_by_customers_copy=orders_by_customers.query("order_status == 'delivered'")
orders_by_customers_copy.head()

```

```python
orders_by_customers_copy['cohort']=orders_by_customers_copy.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')
orders_by_customers_copy['cohort_day']=orders_by_customers_copy.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('D')
```

```python
orders_by_customers_copy['period_number_month']=np.floor((orders_by_customers_copy.order_period_day-orders_by_customers_copy.cohort_day).apply(attrgetter('n'))/30)
```

```python
orders_by_customers_copy.sort_values('period_number_month', ascending=False).head(5)
```

```python
df_cohort=orders_by_customers_copy.groupby(['cohort', 'period_number_month']).agg(n_customers=('customer_unique_id', 'nunique')).reset_index()
df_cohort.sort_values('n_customers', ascending=False)
```

```python
df_cohort_pivot=df_cohort.pivot_table(index='cohort', columns='period_number_month', values='n_customers')
df_cohort_pivot
```

```python
cohort_size=df_cohort_pivot.iloc[:, 0]
retention_matrix=df_cohort_pivot.fillna(0).divide(cohort_size, axis=0)
```

а) продажи handmade товаров (такие товары чаще представлены в единственном экземпляре/небольшом количестве экзмемпляров) 

б) если маркетплейс собирает остатки товаров из других магазинов - то есть товары также представлены в небольшом ассортименте (аналог магазинов familia)

в) Б/У товаров (но по условию задачи товары - новые)



```python


#проверим различие в ценах среди популярных и всех проданных товаров


popular_products=products_rate.query('cnt_ordered>9')

costs_popular=items[items['product_id'].isin(popular_products['product_id'])]['price']
costs_all_items=items['price']

print(f'''
 Медианная стоимость популярных товаров: {round(costs_popular.median(), 2)}
 Медианная стоимость всех товаров: {round(costs_all_items.median(), 2)}

 Средняя стоимость популярных товаров: {round(costs_popular.mean(),2)}
 Средняя стоимость всех товаров: {round(costs_all_items.mean(), 2)}

 Максимальная стоимость популярных товаров: {round(costs_popular.max(),2)}
 Максимальная стоимость всех товаров: {round(costs_all_items.max(), 2)}
 
 Доля продуктов с высокой стоимостью (выше максимальной стоимости среди популярных товаров): {round(items.query('price > 1999.99')['price'].count()/items['price'].count()*100, 2)}%

''')

import scipy.stats as stats

u_stat, p_value = stats.mannwhitneyu(costs_popular, costs_all_items, alternative="two-sided")

print(f"U-статистика: {u_stat:.4f}")
print(f"P-значение: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Статистически значимые различия между группами")
else:
    print("Нет статистически значимых различий")
```

Популярные товары (которые заказали 10 и более раз) не выбиваются по стоимости от общей картины среди всех заказов.



### Выводы на данном этапе

Среди таких барьеров может быть представленный ассортимент на маркетплейсе - ≈95% товаров маркетплейса были заказаны менее 10 раз.

Также возможным барьером к первому и последующим заказам может являться время доставки (если эстимированное время указывается ДО оформления заказа) - при сравнении результатов среди разных регионов видим, что больше заказов производится в регионах с меньшим сроком доставки.

При этом стоит отметить, что задержка редко ведет к отмене заказа - если покупатель готов ждать именно этот товар ≈14 дней (что вероятно дольше рыночного времени), то и задержку данного товара он принимает.
